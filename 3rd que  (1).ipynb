### Data Visualisation

```python
plt.figure(figsize=(20,10))
pd.crosstab(df['Quantity'],df['Category'],df['Profit'],aggfunc='sum').plot(kind='box',stacked=True)
plt.show()
```

```python
df.tail(10)
```

```python
plt.figure(figsize=(20,10))

df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%',startangle=19)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot("Sales",'Profit',data=df,color='m')
plt.show()
```
